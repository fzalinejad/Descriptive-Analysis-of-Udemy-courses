# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading data

```python
df_business = pd.read_csv('3.1-data-sheet-udemy-courses-business-courses.csv')
df_design = pd.read_csv('3.1-data-sheet-udemy-courses-design-courses.csv')
df_music = pd.read_csv('3.1-data-sheet-udemy-courses-music-courses.csv')
df_web = pd.read_csv('3.1-data-sheet-udemy-courses-web-development.csv')
```

```python
frames = [df_business, df_design, df_music, df_web]
df_result = pd.concat(frames)
df_result.reset_index(drop=True, inplace=True)
df_result.head(10)
```

# Q8.Highest and lowest reviews and ratings per subjects

```python
subjects=df_result.subject.unique()
df_bf=df_result.loc[df_result.subject==subjects[0]]
df_gd=df_result.loc[df_result.subject==subjects[1]]
df_mi=df_result.loc[df_result.subject==subjects[2]]
df_wd=df_result.loc[df_result.subject==subjects[3]]

bf_max_review=df_bf[df_bf.num_reviews == df_bf.num_reviews.max()]
bf_min_review=df_bf[df_bf.num_reviews == df_bf.num_reviews.min()]
bf_max_rating=df_bf[df_bf.Rating == df_bf.Rating.max()]
bf_min_rating=df_bf[df_bf.Rating == df_bf.Rating.min()]

gd_max_review=df_gd[df_gd.num_reviews == df_gd.num_reviews.max()]
gd_min_review=df_gd[df_gd.num_reviews == df_gd.num_reviews.min()]
gd_max_rating=df_gd[df_gd.Rating == df_gd.Rating.max()]
gd_min_rating=df_gd[df_gd.Rating == df_gd.Rating.min()]

mi_max_review=df_mi[df_mi.num_reviews == df_mi.num_reviews.max()]
mi_min_review=df_mi[df_mi.num_reviews == df_mi.num_reviews.min()]
mi_max_rating=df_mi[df_mi.Rating == df_mi.Rating.max()]
mi_min_rating=df_mi[df_mi.Rating == df_mi.Rating.min()]

wd_max_review=df_wd[df_wd.num_reviews == df_wd.num_reviews.max()]
wd_min_review=df_wd[df_wd.num_reviews == df_wd.num_reviews.min()]
wd_max_rating=df_wd[df_wd.Rating == df_wd.Rating.max()]
wd_min_rating=df_wd[df_wd.Rating == df_wd.Rating.min()]
```

```python
df_min_max = pd.DataFrame(columns=['subject','max_review_title','max_review','min_review_title','min_review','max_rating_title','max_rating','min_rating_title','min_rating'])
df_min_max.subject=subjects

df_min_max.max_review_title=[bf_max_review.iloc[0,1],gd_max_review.iloc[0,1],mi_max_review.iloc[0,1],wd_max_review.iloc[0,1]]
df_min_max.max_review=[bf_max_review.iloc[0,5],gd_max_review.iloc[0,5],mi_max_review.iloc[0,5],wd_max_review.iloc[0,5]]

df_min_max.min_review_title=[bf_min_review.iloc[0,1],gd_min_review.iloc[0,1],mi_min_review.iloc[0,1],wd_min_review.iloc[0,1]]
df_min_max.min_review=[bf_min_review.iloc[0,5],gd_min_review.iloc[0,5],mi_min_review.iloc[0,5],wd_min_review.iloc[0,5]]

df_min_max.max_rating_title=[bf_max_rating.iloc[0,1],gd_max_rating.iloc[0,1],mi_max_rating.iloc[0,1],wd_max_rating.iloc[0,1]]
df_min_max.max_rating=[bf_max_rating.iloc[0,8],gd_max_rating.iloc[0,8],mi_max_rating.iloc[0,8],wd_max_rating.iloc[0,8]]

df_min_max.min_rating_title=[bf_min_rating.iloc[0,1],gd_min_rating.iloc[0,1],mi_min_rating.iloc[0,1],wd_min_rating.iloc[0,1]]
df_min_max.min_rating=[bf_min_rating.iloc[0,8],gd_min_rating.iloc[0,8],mi_min_rating.iloc[0,8],wd_min_rating.iloc[0,8]]

df_min_max
```

```python
Q1 = df_result[['num_subscribers','num_reviews','num_lectures','content_duration']].quantile(0.25)
Q3 = df_result[['num_subscribers','num_reviews','num_lectures','content_duration']].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

# Removing outliers

```python
df_without_outlier = df_result[~((df_result[['num_subscribers','num_reviews','num_lectures','content_duration']] < (Q1 - 1.5 * IQR)) |(df_result[['num_subscribers','num_reviews','num_lectures','content_duration']] > (Q3 + 1.5 * IQR))).any(axis=1)]
df_without_outlier.head(5)
```

```python
df_without_outlier.info()
```

# Q2.highest mean of reviews and ratings per subjects

```python
mean_reviews=df_without_outlier.groupby(['subject']).num_reviews.mean()
mean_ratings=df_without_outlier.groupby(['subject']).Rating.mean()
```

# Q6. Effect of published date on number of reviews

```python
df_without_outlier['year']=df_without_outlier.published_timestamp.dt.year
df_without_outlier['month']=df_without_outlier.published_timestamp.dt.month
df_without_outlier["year"]=df_without_outlier["year"].values.astype('str')
df_without_outlier["month"]=df_without_outlier["month"].values.astype('str')
df_without_outlier['year_month']=pd.to_datetime(df_without_outlier["year"]+'-'+df_without_outlier["month"])
df_without_outlier
```

```python
year_month=df_without_outlier.groupby(['year_month']).num_reviews.mean()
year_month
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(year_month.index,year_month.values)
ax.set_xlabel('Published date')
ax.set_ylabel('Number of reviews')
ax.set_title('relationship between Published date and Number of reviews')
plt.xticks(rotation=45)
plt.show()
```
